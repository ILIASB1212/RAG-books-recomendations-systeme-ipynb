### CLASSIFICATION

```python

from transformers import pipeline
classes = ["Fiction","Non Fiction"]
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sentence=data.loc[data['categories']!="Fiction","description"].reset_index(drop=True)[30]
sentence
```

```python
x=classifier(sentence,classes)
```

```python
index_score=np.argmax(x['scores'])
print(index_score)
x['labels'][index_score],x['scores'][index_score]
```

```python
x['scores']
```

```python
def get_label(mots,clas):
    prediction=classifier(mots,clas)
    index_score=np.argmax(prediction['labels'])
    x=prediction['labels'][index_score]
    return x
```

```python
get_label(sentence,classes)
```

```python
actual_cats = []
predicted_cats = []

sample_data = data.iloc[0:600]

for i in tqdm(range(len(sample_data))):
    text_description = sample_data.iloc[i]['description']
    
    actual_category = sample_data.iloc[i]['categories']
    
    predicted_result = classifier(text_description, classes)
    predicted_label = predicted_result['labels'][0]

    actual_cats.append(actual_category)
    predicted_cats.append(predicted_label)
    
predictions_df = pd.DataFrame({
    "actual_categories": actual_cats, 
    "predicted_categories": predicted_cats
})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
```
